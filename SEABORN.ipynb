```python
#heatmaps
df=sns.load_dataset('flights')
df.head()
```

```python
df = df.pivot(index='month', columns='year', values='passengers')
print(df.head())
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(df, annot=True, fmt='d')
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(df, annot=True, fmt='d',linecolor='g',linewidths='5')
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(df, annot=True, fmt='d',linecolor='g',linewidths='5',cmap='Blues')
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(df,cbar=False)
```

```python
grid_kws={"height_ratios":(.4,.05),"hspace":.4}
f,(ax,cbar_ax)=plt.subplots(2,gridspec_kw=grid_kws)
ax=sns.heatmap(df,cbar_kws={"orientation":"horizontal"},ax=ax,cbar_ax=cbar_ax)
```
