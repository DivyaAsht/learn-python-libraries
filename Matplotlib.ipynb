```python
tmp_pune=[12,13,14,23,34,45,67,78]
hum_pune=[11,22,33,44,55,66,77,88]

tmp_nag=[98,76,54,65,87,43,32,11]
hum_nag=[99,88,56,40,53,44,21,32]
```

```python
plt.figure(figsize=(8,8))
plt.plot(tmp_pune, hum_pune, 'ro', markersize=10)
plt.show()
```

```python
#barplot
sub=['math','eng','sci','cmp']
marks=[80,87,89,92]
mark=[60,70,80,90]
```

```python
plt.figure(figsize=(5,5))
plt.bar(sub,marks)
plt.bar(sub,mark)
plt.show()
```

```python
sub_len=np.arange(len(sub))
width=0.4

plt.figure(figsize=(5,5))
plt.bar(sub_len,marks,width=width)
plt.bar(sub_len+width,mark,width=width)
plt.show()
```

```python
sub_len=np.arange(len(sub))
width=0.4

plt.figure(figsize=(5,5))
plt.bar(sub_len,marks,width=width,color=col)#b&w
plt.bar(sub_len+width,mark,width=width,color=col,alpha=0.5)
plt.show()
```

```python
#histplot
marks=np.random.randint(0,100,(50))
marks
```

```python
plt.hist(marks)
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(5,5))
plt.hist(marks,bins=bins,color="red")
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(5,5))
plt.hist(marks,bins=bins,color="red",orientation='horizontal')
plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(5,5))
plt.hist(marks,bins=bins,color="red",rwidth=0.6)
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(5,5))
plt.hist(marks,bins=bins,color="red",histtype='step')#bar,barstacked,step,stepfilled
plt.xticks(np.arange(0,100,5))
plt.show()
```
