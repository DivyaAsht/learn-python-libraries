```python
df1=pd.DataFrame({'Roll':[1,2,3,4,5],'Science':[97,80,77,95,87]})
df2=pd.DataFrame({'Roll':[6,2,3,4,8],'Maths':[94,80,71,78,43]})
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
#append
import pandas as pd
df1=pd.DataFrame({'Roll':[1,2,3,4,5],'Maths':[97,80,77,95,87],'Science':[94,80,71,78,43]})
df2=pd.DataFrame({'Roll':[6,7,8,9,10],'Maths':[94,80,71,78,43],'Science':[97,80,77,95,87]})
print(df1)
print(df2)
```

```python
df1.append(df2)   
```

```python
df1.append(df2, ignore_index=True)
```

```python
df1.append(df2, ignore_index=True, sort=True)
```
